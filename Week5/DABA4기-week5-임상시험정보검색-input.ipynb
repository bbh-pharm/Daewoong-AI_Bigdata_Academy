### 페이지번호, 검색어를 넘기면 데이터를 가져오는 함수 만들기

```python
# try, except 를 사용해서 오류가 났을 때 오류대신 메시지만 출력하도록 합니다.
#     try:
#     except Exception as e:
#         return e
def get_page_content(page_no, start_date, end_date):
    base_url ="https://nedrug.mfds.go.kr/searchClinic?"
    params = f"page={page_no}&searchYn=true&approvalStart={start_date}&approvalEnd={end_date}&searchType=ST3&searchKeyword=&approvalDtStart={start_date}&approvalDtEnd={end_date}&clinicStepCode=&examFinish=%EC%8A%B9%EC%9D%B8%EC%99%84%EB%A3%8C&domestic=&gender=&age=&localList=000&localList2=000"
    url = base_url + params
    table = pd.read_html(url, encoding='utf-8')[0]
    
    return table
```

```python
get_page_content(1, '2020-12-03', '2021-06-03')
```

### 반복문을 통해 전체 페이지를 수집

* tqdm은?
    * 오래 걸리는 작업의 진행 상태를 표시합니다.
    * https://tqdm.github.io/

```python
# 페이지마다의 목록을 담아줄 리스트 초기화
page_list = []
page_no = 1

# 반복문을 통해 여러 페이지의 목록 수집하기
while True:
    temp = get_page_content(page_no, '2020-12-03', '2021-06-03')
    if temp.shape[0] > 1:
        page_list.append(temp)
        page_no = page_no + 1
        print(page_no)
    else:
        break
```

### 수집한 데이터 하나의 데이터프레임으로 합치기

```python
df = pd.concat(page_list, ignore_index=True)
df
```

### 파일로 저장하기

```python
file_name = f"임상시험_{today}.csv"
file_name
```

```python
# to_csv 로 파일로 저장하기
df.to_csv(file_name, index=False, encoding='utf-8')
```
