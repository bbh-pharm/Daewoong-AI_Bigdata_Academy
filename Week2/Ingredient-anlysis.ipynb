```python
# 로션제의 함량단위 빈도수  value_counts()로 구하기
df.loc[df['제형명칭'] == '로션제', '함량수치'].value_counts()
```

```python
# 로션제의 "함량수치" 기술통계 값 구하기
df.loc[df['제형명칭'] == '로션제', '함량수치'].describe()
```

```python
# "제형명칭"이 "액제" 이며 "투여경로명" 이 "외용제"인 제품 에서 "함량수치"가 많은 순으로 정렬하기
# 정렬을 해보면 전처리가 좀 더 필요함을 알 수 있습니다.
# df_lq 라는 변수에 지정해서 봅니다.
df_lq = df[(df['제형명칭'] == '액제') & (df['투여경로명'] == '외용제')].sort_values(by='함량수치', ascending=False)
```

## 함량단위

```python
# "함량단위" 의 unique()값만 출력합니다.
df['함량단위'].unique()
```

```python
# df_lq 의 "함량단위"값에 대한 빈도수를 봅니다.
# 어떤 데이터를 전처리 해야할지 볼 때 사용하면 좋습니다.
df_lq['함량단위'].value_counts()
```

```python
# groupby를 사용하여 함량단위 값으로 그룹화를 하여 ["함량수치"]의 describe() 값을 구합니다.
# df_qaunt에 결과를 넣고 "mg", "ml" 에 대한 결과만 가져옵니다.
df_qaunt = df_lq.groupby('함량단위')['함량수치'].describe()
df_qaunt.loc[['mg', 'ml']]
```

## pivot_table과 groupby 

```python
#  pivot_table로 "투여경로명", "제형명칭"을 index 값으로 
# "함량수치"를 value 값으로 하는 "count", "mean" 값을 구합니다.
pd.pivot_table(df, index=['투여경로명', '제형명칭'], values='함량수치', aggfunc=['count', 'mean'])
```

```python
#  groupby로 "투여경로명", "제형명칭"을 index 값으로 
# "함량수치"를 value 값으로 하는 "count", "mean" 값을 구합니다.
df.groupby(['투여경로명', '제형명칭'])['함량수치'].agg(['count', 'mean'])
```

```python
# 제형명칭이 "액제" 인 데이터만 가져와서 df_lq 변수에 담습니다.
df_lq = df[df['제형명칭'] == '액제']
```

```python
# df_lq에서 "단일복합", "투여경로명"을 인덱스 값으로 하고 
# pivot_table 로 "주성분코드"의 count 수와 "함량수치"의 mean 값을 구합니다.
# aggfunc={"주성분코드":"count", "함량수치":"mean"} 을 사용합니다.
pd.pivot_table(df_lq, index=['단일복합', '투여경로명'], aggfunc={"주성분코드":"count", "함량수치":"mean"})
```

```python
# df_lq에서 "단일복합", "투여경로명"을 인덱스 값으로 하고 
# groupby로 주성분코드"의 count 수와 "함량수치"의 mean 값을 구합니다.
# .agg({"주성분코드":"count", "함량수치":"mean"})을 사용합니다.
df_lq.groupby(["단일복합", "투여경로명"]).agg({"주성분코드":"count", "함량수치":"mean"})
```
